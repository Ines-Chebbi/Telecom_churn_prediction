```python
@app.route('/predict', methods=['POST'])
def predict():
    try:
        # Extract form data
        gender = request.form.get('gender')
        SeniorCitizen = request.form.get('SeniorCitizen')
        Partner = request.form.get('Partner')
        Dependents = request.form.get('Dependents')
        tenure = request.form.get('tenure')
        InternetService = request.form.get('InternetService')
        OnlineSecurity = request.form.get('OnlineSecurity')
        OnlineBackup = request.form.get('OnlineBackup')
        TechSupport = request.form.get('TechSupport')
        Contract = request.form.get('Contract')
        PaymentMethod = request.form.get('PaymentMethod')
        MonthlyCharges = request.form.get('MonthlyCharges')
        TotalCharges = request.form.get('TotalCharges')

        # Debug print to verify form data
        print("Form Data Received:", request.form)

        # Process form data and prepare for prediction
        input_data = pd.DataFrame({
            'gender': [gender],
            'SeniorCitizen': [SeniorCitizen],
            'Partner': [Partner],
            'Dependents': [Dependents],
            'tenure': [tenure],
            'InternetService': [InternetService],
            'OnlineSecurity': [OnlineSecurity],
            'OnlineBackup': [OnlineBackup],
            'TechSupport': [TechSupport],
            'Contract': [Contract],
            'PaymentMethod': [PaymentMethod],
            'MonthlyCharges': [MonthlyCharges],
            'TotalCharges': [TotalCharges]
        })

        # Convert categorical features to numerical values
        input_data['gender'] = input_data['gender'].map({'Male': 1, 'Female': 0})
        input_data['SeniorCitizen'] = input_data['SeniorCitizen'].map({'Yes': 1, 'No': 0})
        input_data['Partner'] = input_data['Partner'].map({'Yes': 1, 'No': 0})
        input_data['Dependents'] = input_data['Dependents'].map({'Yes': 1, 'No': 0})
        input_data['InternetService'] = input_data['InternetService'].map({'DSL': 1, 'Fiber optic': 2, 'No': 0})
        input_data['OnlineSecurity'] = input_data['OnlineSecurity'].map({'Yes': 1, 'No': 0, 'No internet service': 2})
        input_data['OnlineBackup'] = input_data['OnlineBackup'].map({'Yes': 1, 'No': 0, 'No internet service': 2})
        input_data['TechSupport'] = input_data['TechSupport'].map({'Yes': 1, 'No': 0, 'No internet service': 2})
        input_data['Contract'] = input_data['Contract'].map({'Month-to-month': 0, 'One year': 1, 'Two year': 2})
        input_data['PaymentMethod'] = input_data['PaymentMethod'].map({
            'Electronic check': 0,
            'Mailed check': 1,
            'Bank transfer (automatic)': 2,
            'Credit card (automatic)': 3
        })

        # Convert tenure, MonthlyCharges, and TotalCharges to numeric
        input_data['tenure'] = pd.to_numeric(input_data['tenure'], errors='coerce')
        input_data['MonthlyCharges'] = pd.to_numeric(input_data['MonthlyCharges'], errors='coerce')
        input_data['TotalCharges'] = pd.to_numeric(input_data['TotalCharges'], errors='coerce')

        # Ensure columns match the training data
        columns = ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'tenure',
                   'InternetService', 'OnlineSecurity', 'OnlineBackup', 'TechSupport',
                   'Contract', 'PaymentMethod', 'MonthlyCharges', 'TotalCharges']
        input_data = input_data[columns]

        # Drop the feature names (convert to numpy array)
        input_array = input_data.to_numpy()

        # Debug prints to ensure data transformation is correct
        print("Processed Input DataFrame:\n", input_data)
        print("Processed Input Array:\n", input_array)

        # Make predictions
        prediction = model.predict(input_array)[0]
        prediction_prob = model.predict_proba(input_array)[0]

        # Debug print to check prediction results
        print("Prediction:", prediction)
        print("Prediction Probability:", prediction_prob)

        # Return the result
        return jsonify({'prediction': 'Churn' if prediction == 1 else 'No Churn',
                        'probability': prediction_prob.tolist()})

    except Exception as e:
        print("Error during prediction:", e)
        return jsonify({'error': str(e)})
```
